## Reading input data

The first step is to load the input vectors and ground truth used for training the SOM.

```python
def read_dataset(data_path, name):
    X = []
    y = []

    with open(data_path + name + "/" + name + ".vec", "r") as vector_file:
        for line in vector_file:
            if line.startswith("$"):
                continue

            X.append(np.array(line.strip().split(" ")[:-1]).astype(float))

    with open(data_path + name + "/" + name + ".cls", "r") as class_file:
        for line in class_file:
            if line.startswith("$"):
                continue

            y.append(line.strip().split(" ")[1])

    return X, y
```

## Loading mnemonic template

The next step is to load the mask which determines which units are active and which ones are deactivated. The image itself can be any black and white image with preferably large white areas which ensure the image is recognizable after the downsampling done by the lower resolution of the node-grid.

Functions used are:

```python
# returns true if the area given by x, y, width and height contains more white pixels than black pixels
# this is used to determine if a unit corresponding to a part of the mnemonic template is active.
def is_active_unit(image, x, y, width, height):
    x0 = min(x, image.shape[1])
    y0 = min(y, image.shape[0])
    x1 = min(x + width, image.shape[1])
    y1 = min(y + height, image.shape[0])

    return (np.sum(image[y0:y1, x0:x1])/image[y0:y1, x0:x1].size) > 0.5

# converts the given image into a matrix with width "som_width" and height: "som_height"
# the resulting matrix contains one element for each unit which is "1" if the unit is active or "0" 
# if the unit is masked out.
def convert_to_active_unit_matrix(image, som_width, som_height):
    unit_width = image.shape[1]/som_width
    unit_height = image.shape[0]/som_height

    active_unit_matrix = np.zeros([som_width, som_height], dtype = int)

    for l in range(som_height):
        for c in range(som_width):
            x = int(c * unit_width)
            y = int(l * unit_height)

            active_unit_matrix[c, l] = (1 if is_active_unit(image, x, y, int(unit_width), int(unit_height)) else 0)

    return active_unit_matrix.T


# converts the given RGB-image to a binary format where all pixels with at least one color at its maximum
# get converted to a "1" and all others become "o"
def convert_to_binary(image):
    return np.amax(image, axis=2)//255
```

```python
# the path of the image used to create the mask for the S
mnemonic_template_path = "mnemonics/stick_figure.png"
# the size of the SOM
som_width = 25
som_height = 25

img = cv2.imread(mnemonic_template_path, cv2.IMREAD_UNCHANGED)
img = convert_to_binary(img)

active_unit_matrix = convert_to_active_unit_matrix(img, som_width, som_height)
```

## Creating the distance table

A lookup table was used to increase the speed at which the SOM is able to deterime the distance between two units.

To store the distance_matrix the class DistanceMatrix was used:

```python
class DistanceMatrix:

    def __init__(self, active_unit_matrix):
        # an index of all active units used for more convinient iteration in some cases
        self.unit_index = []

        for x in range(active_unit_matrix.shape[0]):
            for y in range(active_unit_matrix.shape[1]):
                if active_unit_matrix[x, y] == 1:
                    self.unit_index.append([x, y])

        # the initialized but empty distance matrices for all active units
        self.distance_array = np.full((len(self.unit_index), len(self.unit_index)), np.inf)

    def _get_index(self, x, y):
        for i, elem in enumerate(self.unit_index):
            if x == elem[0] and y == elem[1]:
                return i
        return -1

    # sets the distance between two units given by their id
    def set_distance_by_index(self, i0, i1, value):
        self.distance_array[i0, i1] = value
        self.distance_array[i1, i0] = value

    # returns the distance of two units given by their id
    def get_distance_by_index(self, i0, i1):
        return self.distance_array[i0, i1]

    # sets the distance between two units given by their coordinates
    def set_distance_by_coords(self, x0, y0, x1, y1, value):
        i0 = self._get_index(x0, y0)
        i1 = self._get_index(x1, y1)

        if i0 == -1 or i1 == -1:
            return

        return self.set_distance_by_index(i0, i1, value)

    # returns the distance of two units given by their coordinates
    def get_distance_by_coords(self, x0, y0, x1, y1):
        i0 = self._get_index(x0, y0)
        i1 = self._get_index(x1, y1)

        if i0 == -1 or i1 == -1:
            return np.inf

        return self.get_distance_by_index(i0, i1)

    # returns the amount of active units
    def get_active_unit_count(self):
        return len(self.unit_index)

    # returns a list of active unit coordinates
    def get_active_units(self):
        return self.unit_index
```

For the generation of the distance matrix, an algorithm was used which was derived from the Dijkstra algorithm. The algorithm itself is explained in the comments of the following functions:

```python
# checks if the given coordinate is out of bounds for the given array
def is_out_of_bounds(array, x, y):
    return x < 0 or y < 0 or array.shape[0] <= x or array.shape[1] <= y

# calculates the distance-table for the given unit (given as list containing the x and y coordinates)
def calculate_distances_for_unit(active_unit_matrix, distance_matrix, unit):
    # a matrix that indicates if a unit is not explored yet
    # all inactive units are initialized as already visited
    visited_units = active_unit_matrix != 1 
    
    # a list which will contain the units which will be explored in the current step
    active_units = []
    
    # a list which will contain the units which will be explored in the next step
    # initialized containing the first unit
    next_active_units = [unit]
    
    # sets the first unit as visited
    visited_units[unit[0], unit[1]] = True

    # sets the distance which will be assigned to all units in active_units to 0
    distance = 0 

    # continue until there are no connected units which need to be explored
    while len(next_active_units) > 0:
        # sets the active units for the current step
        active_units = next_active_units
        # clears the active units for the next step
        next_active_units = []

        # continue unitl there are no active units left in the current step
        while len(active_units) > 0:
            # take the first unit from active_units
            current_unit = active_units.pop(0)
            
            # read the coordinates for easier use
            x = current_unit[0]
            y = current_unit[1]

            # set the distance of the current unit in the distance matrix
            distance_matrix.set_distance_by_coords(unit[0], unit[1], x, y, distance)

            # take each non visited active unit in the immidiate neighborhood, add them
            # to the list of active units for the next step and indicate that they are
            # already visited by setting the corresponding value in "visited_units"
            if not is_out_of_bounds(visited_units, x-1, y) and not visited_units[x-1, y]:
                next_active_units.append([x-1, y])
                visited_units[x-1, y] = True
            if not is_out_of_bounds(visited_units, x+1, y) and not visited_units[x+1, y]:
                next_active_units.append([x+1, y])
                visited_units[x+1, y] = True
            if not is_out_of_bounds(visited_units, x, y-1) and not visited_units[x, y-1]:
                next_active_units.append([x, y-1])
                visited_units[x, y-1] = True
            if not is_out_of_bounds(visited_units, x, y+1) and not visited_units[x, y+1]:
                next_active_units.append([x, y+1])
                visited_units[x, y+1] = True

        # increment the distance
        distance += 1

# creates an empty distance matrix and fills it by calculating the distances for every
# active unit in the given active unit matrix
def calculate_distance_matrix(active_unit_matrix):
    distance_matrix = DistanceMatrix(active_unit_matrix)

    for i, unit in enumerate(distance_matrix.get_active_units()):
        calculate_distances_for_unit(active_unit_matrix, distance_matrix, unit)

    return distance_matrix
```

```python
# calculate the distance matrix
distance_matrix = calculate_distance_matrix(active_unit_matrix)
```

## The mnemonic self organizing map

The self organizing map is represented by two objects:

SOMUnit: represents a unit of the SOM <br>
MnemonicSOM: represents the mnemonic SOM itself

```python
class SOMUnit:

    def __init__(self, x, y, input_lenght, initial_learning_rate, learning_rate_decay, initial_radius, radius_decay):
        self.weights = np.random.rand(input_lenght)
        self.x = x
        self.y = y

        self.initial_radius = initial_radius
        self.radius_decay = (radius_decay if radius_decay != 0 else 1)
        self.learning_rate_decay = (learning_rate_decay if learning_rate_decay != 0 else 1)
        self.initial_learning_rate = initial_learning_rate

    # calculates the distance of the current weights to the input vector
    def calculate_distance(self, input_vector):
        return math.sqrt(np.sum((self.weights-input_vector)**2))

    # adjust the weights according to the winning unit
    def adjust_weights(self, input_vector, epoch, distance_matrix, best_unit):
        self.weights += self._learning_rate(epoch) * self._topological_distance(epoch, distance_matrix, best_unit) * (input_vector - self.weights)

    # returns the topological distance corrected for the current epoch
    def _topological_distance(self, epoch, distance_matrix, best_unit):
        return math.exp(-(distance_matrix.get_distance_by_coords(self.x, self.y, best_unit.x, best_unit.y)**2)/(2 * self._neighborhood_size(epoch) **2))

    # returns the neighborhood size corrected for the current epoch
    def _neighborhood_size(self, epoch):
        return self.initial_radius * math.exp(-(epoch)/(self.radius_decay))

    # returns the learning rate corrected for the current epoch
    def _learning_rate(self, epoch):
        return self.initial_learning_rate * math.exp(-(epoch)/(self.learning_rate_decay))

    # returns the current weight vector
    def get_weights(self):
        return self.weights

class MnemonicSOM:

    # initializes the SOM:
    #
    # active_unit_matrix : The active unit matrix used as mask.
    # distance_matrix : The distance matrix for looking up the distances between units
    # input_lenght : The length of an inputvector used to initialize the weight-vectors
    #
    # initial_learning_rate : The initial learning rate at epoch 0.
    # learning_rate_decay : The decay rate of the learning rate used to update the learning rate each epoch.
    # initial_radius : The initial neighborhood radius.
    # radius_decay : The decay of the initial neighborhood radius used to update the neighborhood radius each epoch
    def __init__(self, active_unit_matrix, distance_matrix, input_lenght, initial_learning_rate, learning_rate_decay, initial_radius, radius_decay):
        self.active_unit_matrix = active_unit_matrix
        self.distance_matrix = distance_matrix

        self.initial_radius = initial_radius
        self.radius_decay = (radius_decay if radius_decay != 0 else 1)
        self.learning_rate_decay = (learning_rate_decay if learning_rate_decay != 0 else 1)
        self.initial_learning_rate = initial_learning_rate

        self.input_lenght = input_lenght

        self.units = []

        for x, line in enumerate(active_unit_matrix):
            unit_line = []
            for y, state in enumerate(line):
                if state == 1:
                    unit_line.append(SOMUnit(x, y, input_lenght, 
                        initial_learning_rate = self.initial_learning_rate, 
                        learning_rate_decay = self.learning_rate_decay, 
                        initial_radius = self.initial_radius, 
                        radius_decay = self.radius_decay))
                else:
                    unit_line.append(None)
            self.units.append(unit_line)

    # trains the map using the input X for a given number of epochs
    def train(self, X, epochs):
        for epoch in range(1, epochs+1):

            print("training epoch: ", epoch, "                          ")
            input_set = X.copy()

            while len(input_set) > 0:

                # take random input
                i = rd.randint(0, len(input_set)-1)
                vec = input_set.pop(i)

                # find best unit
                best_unit = self._get_best_unit(vec)

                # adjust weights
                for line in self.units:
                    for unit in line:
                        if unit:
                            unit.adjust_weights(vec, epoch, self.distance_matrix, best_unit)

        print("")
        
    def _get_best_unit(self, input_vector):
        best_unit = None
        best_value = -1

        for line in self.units:
            for unit in line:
                if unit:
                    value = unit.calculate_distance(input_vector)
                    if best_value == -1 or value < best_value:
                        best_unit = unit
                        best_value = value
                    
        return best_unit
    
    # returns the weightlist in a format which can be used by the basic visualizations provided.
    def export_weightlist(self):
        ret = {}

        ret["xdim"] = len(self.units)
        ret["ydim"] = len(self.units[0])
        ret["vec_dim"] = self.input_lenght

        ret["arr"] = []
        for x in range(len(self.units)):
            for y in range(len(self.units[0])):
                if self.units[x][y]:
                    ret["arr"].append(self.units[x][y].get_weights())
                else:
                    ret["arr"].append(np.full(self.input_lenght, -np.inf))

        return ret

    # a debugging function to render the SOM
    def display_som(self, X, y):
        som_image = np.full((len(self.units), len(self.units[0])), 0)

        for vec, label in zip(X, y):
            unit = self._get_best_unit(vec)

            som_image[unit.x, unit.y] = label

        som_image = som_image/np.max(som_image)

        plt.imshow(np.clip(som_image, a_min = 0, a_max = 1))
        plt.show()
```

```python
initial_learning_rate = 2
learning_rate_decay = 5
initial_radius = 2
radius_decay = 10

# initialize the SOM
som = MnemonicSOM(active_unit_matrix, distance_matrix, len(X[0]), 
        initial_learning_rate = initial_learning_rate, 
        learning_rate_decay = learning_rate_decay, 
        initial_radius = initial_radius, 
        radius_decay = radius_decay)

# train the SOM
som.train(X, 10)
```

```python
# display the SOM to check if the training was successful
som.display_som(X, y)
```
